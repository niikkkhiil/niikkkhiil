**6. Define the terms "probability distribution" and "probability density function" (PDF). How does a
PDF differ from a probability mass function (PMF)?**

**Probability Distribution**

A probability distribution describes how the probability of different outcomes is distributed over a set of possible values of a random variable. It provides a function or rule that assigns probabilities to each possible outcome or range of outcomes in a given sample space.

For discrete random variables, this function is called a Probability Mass Function (PMF).

For continuous random variables, this function is called a Probability Density Function (PDF).

**Probability Density Function (PDF)**

A Probability Density Function (PDF) describes the probability distribution of a continuous random variable. It is a function that assigns relative likelihoods to different values within a continuous range. Unlike a PMF, the PDF does not directly give the probability of a specific outcome but the probability of falling within a certain interval.


**Conclusion:**

A probability distribution describes the spread of probabilities for a random variable.

A PDF applies to continuous random variables and gives the relative likelihood of the variable falling within a range.

A PMF applies to discrete random variables and gives the exact probability of each specific outcome.

The Central Limit Theorem (CLT) is one of the most fundamental results in probability theory and statistics. It states that, regardless of the original distribution of a population, the sampling distribution of the sample mean will tend to be approximately normal (Gaussian) if the sample size is sufficiently large.

**8. Compare z-scores and t-scores. When should you use a z-score, and when should a t-score be a
pplied instead?**

Both z-scores and t-scores are used in hypothesis testing and confidence intervals to measure how far a sample statistic is from the population mean, but they are applied in different contexts based on sample size and known/unknown population parameters.

**Z-Scores (Standard Scores)**

Definition: A z-score tells us how many standard deviations a data point is from the population mean.

Purpose: Z-scores are used when we know the population standard deviation (
𝜎
) and need to compare a data point or sample mean to the population mean in terms of standard deviations.

 **T-Scores (T-Statistics)**

Definition: A t-score is similar to a z-score but is used when the sample size is small and/or the population standard deviation is unknown.

Purpose: T-scores are used in place of z-scores when the population standard deviation is unknown and must be estimated from the sample data. The t-distribution accounts for the additional uncertainty from using the sample standard deviation instead of the population standard deviation.

T-Distribution: The t-distribution is similar to the normal distribution but has fatter tails. This accounts for more variability, especially with small sample sizes. As the sample size increases, the t-distribution approaches the normal (z) distribution.

**Summary:**
Use z-scores when you have a large sample size and know the population standard deviation.

Use t-scores when you have a small sample size or the population standard deviation is unknown, and you must estimate it using the sample standard deviation.


9. Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample
size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to
reject the null hypothesis?

 Task: Write Python code to calculate the z-score and p-value for the given data.

Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.


Interpretation:

Significance Level (
𝛼
): 0.05

p-value: 0.096

Since the p-value (0.096) is greater than the significance level
𝛼
=
0.05
, we fail to reject the null hypothesis. This means there is insufficient evidence to conclude that the sample mean is significantly different from the population mean.

```python
import scipy.stats as stats


sample_mean = 105
population_mean = 100
std_dev = 15
sample_size = 25


std_error = std_dev / (sample_size ** 0.5)


z_score = (sample_mean - population_mean) / std_error


p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

z_score, p_value
```
